```python
# Fibonacci Memoization

def fib_calc(n, memo):
#     print(n, memo)
    if memo[n] is not None:
        return memo[n]
    if n == 1 or n == 2:
        result = 1
    else:
        result = fib_calc(n-2, memo) + fib_calc(n-1, memo)
    memo[n] = result 
#     print(n, memo)
    return result 

def fib(n):
    memo = [None] * (n+1)
    return fib_calc(n, memo)

fib(3000)
# or can do this: fib_calc(3000, [None]*(3000+1))
```

```python
# Fibonacci Memoization 2

def fib_calc2(n, memo):
    if (n in memo):
        return memo[n]
    if n == 1 or n == 2:
        result = 1
    else:
        result = fib_calc2(n-2, memo) + fib_calc2(n-1, memo)
    memo[n] = result 
    return result 

def fib2(n):
    memo = {}
    return fib_calc2(n, memo)

fib2(3000)
```

```python
# Fibonacci bottom-up 

def fib3(n):
    bot_up = [None] * (n+1)
    bot_up[1] = 1
    bot_up[2] = 1
    for i in range(3, n+1):
        bot_up[i] = bot_up[i-2] + bot_up[i-1]
    return bot_up[n]

fib3(3000)
```
